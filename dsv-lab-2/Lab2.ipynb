## EXERCISE 2 - Operations on DataFrames

### Vectorized arithmetic

Starting from the temperature DataFrame (`city_temps`), calculate the temperatures in Celsius from those expressed in Fahrenheit.

The conversion formula is:

$$C = \frac{5}{9} \cdot ( F - 32  )$$

```python
temperature_df=city_temps.drop(["name"],axis=1)
temperature_df["temperature_C"] = temperature_df["temperature_F"].map(lambda t: 5/9*(t-32))
temperature_df
```

Using the same conversion, add a column to the temperature DataFrame with the temperatures in Celsius. Would rounding be useful?

Hint: you can assign a new column using `df['new_column'] = values`

```python
city_temps["temperature_C"] = city_temps["temperature_F"].map(lambda t: round(5/9*(t-32),2))
city_temps
```

Calculate (and add as a new column *rate*) the murder rate per 100,000 people in the murders DataFrame and show the updated DataFrame

```python
murders["Rate"] = murders["Deaths"]*100000/murders["Population"]
murders.sort_values(by="Rate")
```

What is the average murder rate?

```python
print(round(murders["Rate"].mean(),6))
```

You can use the `.quantile()` method or `.describe()` to calculate the five-number summary (minimum, 1st quartile, median, 3rd quartile, maximum) of a column's data.
What are the 5 numbers that summarize the *rate* column?

```python
min = murders["Rate"].min()
quartile_1 = murders["Rate"].quantile(0.25)
median = murders["Rate"].median()
quartile_3 = murders["Rate"].quantile(0.75)
max = murders["Rate"].max()
```

## Categorical Binning

The `pd.cut()` function allows you to divide numerical values into intervals (given the boundaries, with upper bound included by default) and assign a label to each interval. The result is a categorical Series.

```python
## Example of using pd.cut()
bins = [0, 2e6, 1e7, 5e7]
labels = ["small", "medium", "large"]
pd.cut(murders["Population"], bins=bins, labels=labels, ordered=True).value_counts(sort=False)
```

Assign a level (low, medium, high) to the murder rate corresponding to the intervals (0,10], (10,20], (20,40], and add this classification as a new column *"Rate class"*.

```python
rate_bins = [0,10,20,40]
rate_labels = ["low","medium","high"]
murders["Rate class"] = pd.cut(murders["Rate"],bins=rate_bins,labels=rate_labels)
murders
```

Calculate the frequencies of the different ratings:

```python
murders["Rate class"].value_counts().sort_index()
```

Show the names of the states with a high rate 

```python
mask_rate = murders["Rate class"]=="high"
murders[mask_rate]
```

Show all the details of the states with high rate except the "Rate class" column itself sorted by descending rate

```python
murders[mask_rate].iloc[:,0:6].sort_values(by="Rate",ascending=False)
```

### Grouping

Show the average death rate per region

```python
murders.groupby("Region")["Rate"].mean()
```

Use `agg()` to get a DataFrame instead of a Series

```python
rate_per_region = murders.groupby("Region").agg({"Rate":["mean"]})
rate_per_region.droplevel(0,1).reset_index().rename({"mean":"Average Death Ratio"},axis=1)
```
